# Ejercicios


1. Crea una serie con numeros correlativos entre 1 y 12 inclusive (puedes hacerlo como lista o diccionario): 


```python
import pandas as pd
```

```python
s = pd.Series(range(1,13))
print(s)
```

    - Cambia el indice de la serie anteriormente creada para que comience con el numero 20.

```python
s.index = range(20,32)
print(s)
```

```python
# Puedo hacerlo si vuelvo a crear la serie entera.

s = pd.Series(range(1,13), index=range(20,32))

print(s)
```

    - Nuevamente cambia el indice de la serie, esta vez con una lista que haga referencia a los meses del año (tenemos 12 elementos como la cantidad de meses del año).


```python
meses = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiempre', 'octubre', 'noviembre', 'diciembre']
s.index = meses
print(s)
```

    - Muestra los valores que estan en la posicion 3 y 8 de mi serie.

```python
print(f"En la posición 3 está: {s[3]}")
print(f"En la posición 8 está: {s[8]}")
print(f"En las posiciones 3 y 8 tenemos: {s[[3, 8]]}")
```

    - Muestra los valores que estan etiquetados en enero y agosto.

```python
print(f"En la etiqueta de enero está: {s['enero']}")
print(f"En la etiqueta de agosto está: {s['agosto']}")
print(f"Los valores de enero y agosto son {s[['enero', 'agosto']]}")
```

    - Muestra sus valores, indices, forma, tamaño y el tipo de elementos de la serie.

```python
print(f"Los valores de la Serie son: {s.values}")
print(f"Los índices de la Serie son: {s.index}")
print(f"El tamaño (número total de elementos) de la Serie es: {s.size}")
print(f"La forma de la Serie es: {s.shape}")
print(f"El tipo de elementos de la Serie es: {s.dtype}")
```
