1. Las columnas land_area, armed_forces_size, co2-emissions, gdp, population, urban_population, latitude, longitude, density, cpi son de tipo object pero en realidad deberían ser numéricas. Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde.

```python
def cambiar_dato(cadena):

    try:
        return float(cadena.replace(",", ""))

    except:
        return np.nan
```

```python
lista_columnas = ["land_area", "armed_forces_size", "co2-emissions", "gdp", "population", "urban_population", "latitude", "longitude", "density", "cpi"]
for col in lista_columnas:
    df_w[col] = df_w[col].apply(cambiar_dato)
```

5. ¿Cuál es la tasa de impuestos total máxima por densidad de población por continente? Para solucionar este ejercicio puedes:

- PRIMERO --> calcula la tasa de impuestos total para cada continente. La tasa de impuestos total se encuentra en la columna "total_tax_rate".

- SEGUNDO -->  calcula la densidad de población para cada continente. La densidad de población se encuentra en la columna "density".

- TERCERO -->  divide la tasa de impuestos total entre la densidad de población para obtener la tasa de impuestos total por persona por kilómetro cuadrado.

- CUARTO --> Encuentra el país que tiene el valor más alto en esta nueva columna calculada.

```python
tasa_impuestos_continente = df_w.groupby("continente")["total_tax_rate"].mean().reset_index()
tasa_impuestos_continente
```

```python
tasa_densidad_continente = df_w.groupby("continente")["density"].mean().reset_index()
tasa_densidad_continente
```

```python
df_tasa_cont = tasa_impuestos_continente.merge(tasa_densidad_continente, on = "continente")
df_tasa_cont
```

```python
def calcular_impuesto_persona(t_impuesto, t_densidad):

    resultado = t_impuesto / t_densidad
    return resultado

df_tasa_cont["impuesto_persona"] = df_tasa_cont.apply(lambda x: calcular_impuesto_persona(x["total_tax_rate"], x["density"]), axis = 1)
```

8. ¿Cuál es la tasa de maternidad promedio por tasa de impuestos total? Para resolver el ejercicio deberás:

- PRIMERO --> calcula la tasa de maternidad promedio para cada continente. Puedes usar la columna "maternal_mortality_ratio" que representa la tasa de mortalidad materna.

- SEGUNDO --> calcula la tasa de impuestos total promedio para cada continente. Puedes usar la columna "total_tax_rate" que representa la tasa de impuestos total en cada continente.

- TERCERO --> divide la tasa de maternidad promedio por la tasa de impuestos total promedio para obtener la relación entre ambas tasas.

- CUARTO --> encuentra el país con la tasa de maternidad más alta por punto porcentual de tasa de impuestos total.

```python
df_tasa_mort_mat = df_w.groupby(["total_tax_rate","country"])["maternal_mortality_ratio"].mean().reset_index()
df_tasa_mort_mat["division_punto_porcentual"] = df_tasa_mort_mat["maternal_mortality_ratio"] / df_tasa_mort_mat["total_tax_rate"]
df_tasa_mort_mat
# cuanto afecta que aumente el total de impuestos conforme al ratio de maternidad materna (relacion)
```

```python
valor_maximo_porcentual = df_tasa_mort_mat["division_punto_porcentual"].max()
condicion =  df_tasa_mort_mat["division_punto_porcentual"] == valor_maximo_porcentual

df_respuesta = df_tasa_mort_mat[condicion]
df_respuesta
```

```python
valor_maximo_porcentual = df_tasa_mort_mat["division_punto_porcentual"].max()
registro_respuesta =  df_tasa_mort_mat[df_tasa_mort_mat["division_punto_porcentual"] == valor_maximo_porcentual]
registro_respuesta
```
